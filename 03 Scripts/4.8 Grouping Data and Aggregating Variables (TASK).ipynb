```python
# Grouping the Data
ords_prods_merge.groupby('product_name')
```

```python
# Finding the aggregated mean 
ords_prods_merge.groupby('department_id').agg({'amount_of_orders_made': ['mean']})
```

The results for the entire dataframe show each department_id as opposed to the subset where we only saw 8 of the 21 department_id's. 

```python
# Creating a loyalty flag
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['amount_of_orders_made'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts (dropna = False)
```

```python
# Spending habits of the three types of customers
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Creating a spending flag
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
ords_prods_merge = ords_prods_merge.drop(columns = ['max_order_final'])
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'low_spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'high_spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Creating an order frequency flag
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'non_frequent_customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'regular_customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'frequent_customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge = ords_prods_merge.drop(columns = ['spending flag'])
```
