# 4.7 Exercise

```python
# Create a subset
df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 4.7 Task

```python
# Orders day of week value count
ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

```python
# Creating new column for busiest and least busy days of week
results = []

for value in ords_prods_merged["order_day_of_week"]:
  if value == 0 or value == 1:
    results.append("Busiest days")
  elif value == 4 or value == 3:
    results.append("Slowest days")
  else:
    results.append("Regularly busy")
```

```python
results
```

```python
ords_prods_merged['busiest days'] = results
```

```python
# Check Values of new column
ords_prods_merged['busiest days'].value_counts(dropna = False)
```

```python
# Creating a new column for busiest period of the day
ords_prods_merged.head()
```

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
results_label_hour = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
    results_label_hour.append("Fewest Orders")  
  elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22: 
    results_label_hour.append("Average Orders")  
  elif value == 23 or value == 6 or value == 0 or value == 1 or value == 5 or value == 2 or value == 4 or value == 3:
    results_label_hour.append("Most Orders")
```

```python
results_label_hour
```

```python
ords_prods_merged['Busiest Order Periods'] = results_label_hour
```

```python
# Checking frequency and values for new column
ords_prods_merged['Busiest Order Periods'].value_counts(dropna = False)
```
